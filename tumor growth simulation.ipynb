```python
# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  basic setup
time_steps = 100             # how long to simulate
cells = np.zeros(time_steps) # empty array to store cell count over time
cells[0] = 10                # begn: 10 tumor cells

# parameters
growth_rate = 0.2            # how fast cells divide
death_rate = 0.05            # natural death rate
nutrient_limit = 1000        # maximum cells supported 
```

```python
#for loop:
for t in range(1, time_steps):  #startng at time=1
    growth = growth_rate * cells[t-1] * (1 - cells[t-1] / nutrient_limit) # growth= rate of grrowth* no. of cells * nutient left
    death = death_rate * cells[t-1]
    cells[t] = cells[t-1] + growth - death 
    
```

```python
# Plot 
plt.figure(figsize=(8,5))
plt.plot(cells, color='purple', linewidth=2)
plt.xlabel('Time (arbitrary units)')
plt.ylabel('Number of Tumor Cells')
plt.title('Simulated Tumor Growth Over Time')
plt.grid(True)
plt.show()
```
